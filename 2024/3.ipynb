```python
DAY = notebook.split('.')[0]
with open(f'{DAY}.txt','r') as f:
    INPUT = f.read()
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import re
```

```python
matches = re.findall(r"mul\(([0-9]+),([0-9]+)\)",INPUT)
```

```python
matches
```

```python
s=0
for m1,m2 in matches:
    s+= int(m1)*int(m2)
```

```python
s
```

```python
matches = re.findall(r"mul\(([0-9]+),([0-9]+)\)|(do)\(\)|(don)'t\(\)",INPUT)
```

```python
matches
```

```python
enable=True
s=0
for m1,m2,do,dont in matches:
    if enable and not do and not dont:
        s+= int(m1)*int(m2)
    if dont:
        enable=False
    if do:
        enable=True
```
